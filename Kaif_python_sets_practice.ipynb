### Q7

### Q12
Students in two classes:
```python
class_A = {'Aman','Riya','Karan','Neha'}
class_B = {'Neha','Simran','Aman'}
```

````python
# write your code here
class_A = {'Aman','Riya','Karan','Neha'}
class_B = {'Neha','Simran','Aman'}
print("Union:- ",class_A|class_B)
print("Intersection:- ",class_A&class_B)
print("Difference:- ",class_A-class_B)
print("Symmetric Difference:- ",class_A^class_B)
````
